```python
import glob
import pandas as pd
import numpy as np
import matplotlib


files = glob.glob("/home/train_data/img_csv/*.csv")
```

```python
data_arr = np.zeros((432, 300000))
count =[]
for index, file in enumerate(files):
    data = pd.read_csv(file) 
    data = data.values
    data = np.delete(data, 0, axis = 1)
    count.append(data.shape[0])
    if data.shape[0] > 100000:
        data = np.delete(data, 100000, axis = 0)
    
    data = data.reshape((300000,))
    data_arr[index] = data
    
```

```python
files_test = glob.glob("/home/test_data/img_csv/*.csv")
count1 = []
data_arr1 = np.zeros((108, 300000))
for index1, file1 in enumerate(files_test):
    data1 = pd.read_csv(file1) 
    data1 = data1.values
    data1 = np.delete(data1, 0, axis = 1)
    count1.append(data1.shape[0])
    if data1.shape[0] > 100000:
        data1 = np.delete(data1, 100000, axis = 0)
    
    data1 = data1.reshape((300000,))
    data_arr1[index1] = data1
```

```python
data_som = np.append(data_arr, data_arr1, axis = 0)
```
