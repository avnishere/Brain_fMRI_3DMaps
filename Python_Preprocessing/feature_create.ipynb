```python
import glob
file_list = glob.glob("/home/train_data/images/*.nii.gz")
list_of_count =[0]
```

```python
import numpy as np
import nibabel as nib

for files in file_list:
    img = nib.load(files)
    data_mat = img.get_data()
    feature = np.transpose(np.nonzero(data_mat))
    maximum_local = feature.shape[0]
    #print(maximum_local)
    list_of_count.append(maximum_local)
    
val = max(list_of_count)
```

```python
 new = np.array(list_of_count)
```
